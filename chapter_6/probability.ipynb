```python
import random, enum
from matplotlib import pyplot as plt
```

```python
class Kid(enum.Enum):
	"""
	Enum class to represent the kids in the game.
	"""

	BOY = 0
	GIRL = 1
```

```python
def random_kid():
	return random.choice(list(Kid))
```

```python
both_girls = 0
older_girl = 0
either_girl = 0

random.seed(0)  # For reproducibility
```

```python
for _ in range(100000):
	younger = random_kid()
	older = random_kid()
	if older == Kid.GIRL:
		older_girl += 1

	if older == Kid.GIRL and younger == Kid.GIRL:
		both_girls += 1

	if older == Kid.GIRL or younger == Kid.GIRL:
		either_girl += 1

print(f'P(both|older): {both_girls / older_girl}')
print(f'P(both|either): {both_girls / either_girl}')
```

```python
# unifdorm distribution
def uniform_pdf(x: float):
	"""
	PDF for a uniform distribution.
	"""
	return 1.0 if 0 <= x <= 1 else 0.0
```

```python
# Uniform cumulative distribution function
def uniform_cdf(x: float):
	"""
	CDF for a uniform distribution.
	"""
	if x < 0:
		return 0.0
	elif x <= 1:
		return x
	else:
		return 1.0
```

```python
# use seaborn to plot the PDF and CDF
x = [i / 10.0 for i in range(0, 11)]
pdf = [uniform_pdf(i) for i in x]
cdf = [uniform_cdf(i) for i in x]
fig, ax = plt.subplots(2, 1, figsize=(8, 6))
ax[0].plot(x, pdf, label='PDF', color='blue')
ax[0].set_title('Uniform PDF')
ax[0].set_xlabel('x')
ax[0].set_ylabel('Probability Density')
ax[0].legend()
ax[1].plot(x, cdf, label='CDF', color='orange')
ax[1].set_title('Uniform CDF')
ax[1].set_xlabel('x')
ax[1].set_ylabel('Cumulative Probability')
ax[1].legend()
plt.tight_layout()
plt.show()
```

```python
import math

SQRT2PI = math.sqrt(2 * math.pi)


def normal_pdf(x: float, mu: float = 0, sigma: float = 1):
	"""
	PDF for a normal distribution.
	"""
	return math.exp(-0.5 * ((x - mu) / sigma) ** 2) / (sigma * SQRT2PI)
```
